```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
df = pd.read_csv("/content/KAG_conversion_data.csv")
df.head()
```

```python
df.info()
print("Missing values:\n", df.isnull().sum())
df.describe()
```

```python
print("Unique values in 'age':", df['age'].unique())
print("Unique values in 'gender':", df['gender'].unique())
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

```python
df.info()
```

```python
#Click-Through Rate (CTR)
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100
#Cost Per Click (CPC)
df['CPC'] = df['Spent'] / df['Clicks']
#Conversion Rate
df['Conversion Rate'] = (df['Approved_Conversion'] / df['Clicks']) * 100
```

```python
df.head()
```

```python
# Visualize outliers in 'Spent'
sns.boxplot(df['Spent'])
plt.title('Boxplot of Spent')
plt.show()

# Optionally, cap outliers at 99th percentile
q = df['Spent'].quantile(0.99)
df['Spent'] = np.where(df['Spent'] > q, q, df['Spent'])
```

```python
metrics = ['Spent', 'Clicks', 'Impressions']
for metric in metrics:
    sns.histplot(df[metric], kde=True)
    plt.title(f"Distribution of {metric}")
    plt.show()
```

```python
sns.pairplot(df[['Impressions', 'Clicks', 'Spent', 'Approved_Conversion']])
plt.show()

correlation = df[['Impressions', 'Clicks', 'Spent', 'Approved_Conversion']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Gender-wise clicks and conversions
sns.boxplot(x='gender', y='Clicks', data=df)
plt.title("Gender vs Clicks")
plt.show()
```

```python
# Age group
age_grouped = df.groupby('age').agg({'Clicks': 'sum', 'Approved_Conversion': 'sum'})
age_grouped.plot(kind='bar')
plt.title("Performance by Age Group")
plt.show()
```

```python
campaign_performance = df.groupby('xyz_campaign_id').agg({
    'Spent': 'sum',
    'Approved_Conversion': 'sum'
}).reset_index()
```

```python
campaign_performance['Cost per Conversion'] = (
    campaign_performance['Spent'] / campaign_performance['Approved_Conversion']
)
campaign_performance.sort_values(by='Cost per Conversion', inplace=True)
```

```python
sns.barplot(x='xyz_campaign_id', y='Cost per Conversion', data=campaign_performance)
plt.title("Campaign Performance (Cost per Conversion)")
plt.xlabel("Campaign ID")
plt.ylabel("Cost per Conversion")
plt.show()
```

# Audience Segmentation Analysis

```python
age_performance = df.groupby('age').agg({
    'Clicks': 'sum',
    'Approved_Conversion': 'sum',
    'Spent': 'sum'
}).reset_index()

age_performance['Cost per Conversion'] = (
    age_performance['Spent'] / age_performance['Approved_Conversion']
)
print(age_performance)
```

```python
gender_performance = df.groupby('gender').agg({
    'Clicks': 'sum',
    'Approved_Conversion': 'sum',
    'Spent': 'sum'
}).reset_index()

gender_performance['Cost per Conversion'] = (
    gender_performance['Spent'] / gender_performance['Approved_Conversion']
)
print(gender_performance)
```

```python
interest_performance = df.groupby('interest').agg({
    'Clicks': 'sum',
    'Approved_Conversion': 'sum',
    'Spent': 'sum'
}).reset_index()

interest_performance['Cost per Conversion'] = (
    interest_performance['Spent'] / interest_performance['Approved_Conversion']
)
interest_performance.sort_values(by='Cost per Conversion', inplace=True)
print(interest_performance.head())
```

```python
sns.barplot(x='age', y='Cost per Conversion', data=age_performance)
plt.title("Cost per Conversion by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Cost per Conversion")
plt.show()
```

```python
sns.barplot(x='gender', y='Cost per Conversion', data=gender_performance)
plt.title("Cost per Conversion by Gender")
plt.xlabel("Gender")
plt.ylabel("Cost per Conversion")
plt.show()
```

# **Campaign Performance Evaluation**

```python
campaign_performance = df.groupby('xyz_campaign_id').agg({
    'Impressions': 'sum',
    'Clicks': 'sum',
    'Approved_Conversion': 'sum',
    'Spent': 'sum'
}).reset_index()

campaign_performance['CTR (%)'] = (campaign_performance['Clicks'] / campaign_performance['Impressions']) * 100
campaign_performance['Cost per Click'] = campaign_performance['Spent'] / campaign_performance['Clicks']
campaign_performance['Cost per Conversion'] = campaign_performance['Spent'] / campaign_performance['Approved_Conversion']

print(campaign_performance)
```

```python
campaign_performance.sort_values(by='Cost per Conversion', inplace=True)
print(campaign_performance[['xyz_campaign_id', 'Cost per Conversion', 'CTR (%)']])
```

```python
sns.barplot(x='xyz_campaign_id', y='CTR (%)', data=campaign_performance)
plt.title("Click-Through Rate (CTR) by Campaign")
plt.xlabel("Campaign ID")
plt.ylabel("CTR (%)")
plt.show()
```

```python
sns.barplot(x='xyz_campaign_id', y='Cost per Conversion', data=campaign_performance)
plt.title("Cost per Conversion by Campaign")
plt.xlabel("Campaign ID")
plt.ylabel("Cost per Conversion")
plt.show()
```

# **ROI Analysis**

```python
ARC = 500  # Average Revenue per Conversion (assumed value)
campaign_performance['Revenue Generated'] = campaign_performance['Approved_Conversion'] * ARC
campaign_performance['ROI (%)'] = (
    (campaign_performance['Revenue Generated'] - campaign_performance['Spent']) / campaign_performance['Spent']
) * 100

print(campaign_performance[['xyz_campaign_id', 'Revenue Generated', 'Spent', 'ROI (%)']])
```

```python
high_roi_campaigns = campaign_performance[campaign_performance['ROI (%)'] > 0]
low_roi_campaigns = campaign_performance[campaign_performance['ROI (%)'] <= 0]

print("High ROI Campaigns:\n", high_roi_campaigns)
print("Low ROI Campaigns:\n", low_roi_campaigns)
```

```python
sns.barplot(x='xyz_campaign_id', y='ROI (%)', data=campaign_performance)
plt.title("ROI by Campaign")
plt.xlabel("Campaign ID")
plt.ylabel("ROI (%)")
plt.show()
```

```python
# Visualize Age Group vs Cost per Conversion
sns.barplot(x='age', y='Cost per Conversion', data=age_performance)
plt.title("Cost per Conversion by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Cost per Conversion")
plt.show()

# Visualize Campaign ROI
sns.barplot(x='xyz_campaign_id', y='ROI (%)', data=campaign_performance)
plt.title("ROI by Campaign")
plt.xlabel("Campaign ID")
plt.ylabel("ROI (%)")
plt.show()
```
